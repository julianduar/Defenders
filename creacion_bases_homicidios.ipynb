```python
#Me quedo con los archivos xlsx
files_xlsx = [f for f in files if f[-4:] == 'xlsx' ]
files_xlsx
```

```python
# data frame vacío 
df = pd.DataFrame()
```

```python
## concateno todos los archivos xlsx en el data frame 
for f in files_xlsx:
    data=pd.read_excel(f, engine='openpyxl')
    df = df.append(data)
```

```python
df
```

```python
df = df[df["CODIGO DANE"].notna()] #limpio el contenido basura
```

```python
df = df.reset_index()
```

```python
df
```

```python
del df['index']
```

```python
df['FECHA HECHO']
```

```python
#creación de variables 
df['mes'] = df['FECHA HECHO'].dt.month
df['año'] = df['FECHA HECHO'].dt.year
df['dia'] = df['FECHA HECHO'].dt.day
```

```python
df
```

```python
#variable semestre
df['semestre']=2
```

```python
df.loc[df['mes'] <= 6, 'semestre'] = 1
```

```python
# Estandarización de la variable departamento
df.loc[(df['MUNICIPIO'] =='BOGOTÁ D.C. (CT)') , 'DEPARTAMENTO'] = 'BOGOTA'
df.loc[(df['DEPARTAMENTO'] =='GUAJIRA') , 'DEPARTAMENTO'] = 'LA GUAJIRA'
df.loc[(df['DEPARTAMENTO'] =='VALLE') , 'DEPARTAMENTO'] = 'VALLE DEL CAUCA'
```

```python
df['DEPARTAMENTO']= df['DEPARTAMENTO'].apply(unidecode.unidecode)
```

```python
#Agrupo por departamento
data_by_departamento = df.groupby(['DEPARTAMENTO'])
```

```python
#Sumo el número de homicidios por departamento
data_by_departamento_count = data_by_departamento['CANTIDAD'].agg('sum')
```

```python
data_by_departamento_count
```

```python
#Agrupo por departamento y año
data_by_departamento_year = df.groupby(['DEPARTAMENTO','año'])
```

```python
#Sumo el número de homicidios por departamento y año
data_by_departamento_year_count = data_by_departamento_year['CANTIDAD'].agg('sum')
```

```python
#Paso la información a un data frame
homicidios_departamento=data_by_departamento_year_count.to_frame()
```

```python
homicidios_departamento=homicidios_departamento.reset_index()
```

```python
homicidios_departamento.rename(columns={'DEPARTAMENTO':'departamento'}, inplace=True)
```

```python
homicidios_departamento
```

Importo la base de datos del DANE sobre las proyecciones de la población con los años que me interesan. Con el fin de hallar la tasa de homicidios promedio en el periodo 2016-2020

```python
df_poblacion=pd.read_stata("poblacion_DANE_departamentos.dta")
```

```python
# Estandarización de la variable departamento
df_poblacion.loc[(df_poblacion['Departamento'] =='N. DE SANTANDER') , 'Departamento'] = 'NORTE DE SANTANDER'
df_poblacion.loc[(df_poblacion['Departamento'] =='NARIÑO') , 'Departamento'] = 'NARINO'
```

```python
#renombro las columnas
df_poblacion.rename(columns={'Departamento':'departamento', 'Año':'año'}, inplace=True)
```

```python
df_poblacion[:33]
```

```python
# Realizo el merge de la base homicidios y la base del DANE
tasa_homicidios_departamento=pd.merge(homicidios_departamento,df_poblacion, how='outer',on = ['departamento', 'año'])
```

```python
tasa_homicidios_departamento
```

```python
#Calculo la tasa de homicidios por cada 100 mil habitantes
tasa_homicidios_departamento['tasa_de_homicidios']=(tasa_homicidios_departamento['CANTIDAD']/
                                                    tasa_homicidios_departamento['PoblacióntotalDANE'])*100000
```

```python
tasa_homicidios_departamento
```

```python
#agrupo por departamento
tasa_homicidios_by_departamento= tasa_homicidios_departamento.groupby(['departamento'])
```

```python
#Hallo la tase de homicidios promedio del periodo 2016-2020 para cada departamento
tasa_homicidios_by_departamento_mean= tasa_homicidios_by_departamento['tasa_de_homicidios'].agg('mean')
```

```python
#Lo convierto en un data frame
tasa_homicidios_by_departamento_mean=tasa_homicidios_by_departamento_mean.to_frame()
```

```python
tasa_homicidios_by_departamento_mean=tasa_homicidios_by_departamento_mean.reset_index()
```
