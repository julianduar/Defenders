Al revisar varias de las tablas me percato que la mayoría de las tablas están repetidas, exceptuando las tablas en la posición 2 y 26, estás dos son únicas. Por lo tanto, me interesa recorrer los pares de la variable tables2.

```python
#Se crea un Data Frame vacío con el fin de unir verticalmente todas las tablas en la variable tables 
df2 = pd.DataFrame()


for i in range(0,27,2):
    dftemp=tables2[i].df #Se obtiene el data frame i de la variable tables2 
    dftemp=dftemp.replace('', np.nan)#Se reemplazan las observaciones que no contienen nada con nan
    dftemp=dftemp.drop(0) # Se borra la fila en la posición 0
    dftemp= dftemp.dropna(how='all') #Se borran las filas que contengan solo nan
    colval=dftemp.columns.values #Se guarda el nombre actual de las columnas
    print(colval)
    #creo una lista con los nombres de las columnas que me interesaría tener
    colnames=['N','fecha','departamento','municipio','victima', 'organizacion', 'tipo de lider', 'presunto responsable']
    dic_dftemp={} #Defino un diccionario temporal para ir cambiando el nombre de las columnas 
    for i in range(8):
        if len(colval)==8: #La posición cero es un caso especial
            dic_dftemp[colval[i]]=colnames[i]
        else:
            dic_dftemp[colval[i+1]]=colnames[i]
    dftemp=dftemp.rename(columns=dic_dftemp) #Renombro las columnas
    dftemp=dftemp.replace(r'\n',  ' ', regex=True) #Reemplazo los enters por espacios
    df2=df2.append(dftemp)
```
